### Ultrasonic sensor test example ###

```python
import time
from IPython.display import clear_output

sonic = overlay.sonic_mm_0.mmio   # Create an object for MMIO access to the ultrasonic sensor
sonic_base_addr = sonic.base_addr # Assign the base address of the ultrasonic sensor object

# control register (0x00) bit mapping
# bit 0 -> sensor 1 enable
# bit 1 -> sensor 2 enable
# bit 2 -> sensor 3 enable
# bit 3 -> sensor 4 enable
# bit 4 -> sensor 5 enable
#
# example values
# 0x01 : sensor 1 only
# 0x02 : sensor 2 only
# 0x04 : sensor 3 only
# 0x08 : sensor 4 only
# 0x10 : sensor 5 only
# 0x03 : sensor 1 + 2
# 0x0C : sensor 3 + 4
# 0x1F : all 5 sensors enable

# 5개 센서 enable
sonic.write(0x00, 0x1F)

# 75 MHz clock 기준
CLK_FREQ = 75_000_000
SOUND_SPEED_CM_S = 34300

def ticks_to_cm(count):
    return count * SOUND_SPEED_CM_S / (2 * CLK_FREQ)

try:
    while True:
        ctrl = sonic.read(0x00)
        val1 = sonic.read(0x04)
        val2 = sonic.read(0x08)
        val3 = sonic.read(0x0C)
        val4 = sonic.read(0x10)
        val5 = sonic.read(0x14)

        dist1 = ticks_to_cm(val1)
        dist2 = ticks_to_cm(val2)
        dist3 = ticks_to_cm(val3)
        dist4 = ticks_to_cm(val4)
        dist5 = ticks_to_cm(val5)

        print("ctrl :", ctrl)
        print(f"val1 : {val1:10d}  ->  {dist1:7.2f} cm")
        print(f"val2 : {val2:10d}  ->  {dist2:7.2f} cm")
        print(f"val3 : {val3:10d}  ->  {dist3:7.2f} cm")
        print(f"val4 : {val4:10d}  ->  {dist4:7.2f} cm")
        print(f"val5 : {val5:10d}  ->  {dist5:7.2f} cm")

        time.sleep(0.1)
        clear_output(wait=True)

except KeyboardInterrupt:
    print("Interrupt detected, shutting down gracefully...")

print("done")
```
